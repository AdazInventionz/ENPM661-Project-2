```python
#Imported libraries
import numpy as np
import cv2 as cv
import time
import heapq as hq




def backtrack():
    
    p = len(closedNodes) - 1
    b = closedNodes[p]
    #solution.append(b)
    
    while b[2] > 0:
        
        for i in range(p, 0, -1):
            
            if closedNodes[i][1] == b[2]:
                
                b = closedNodes[i]
                p = i
                solution.insert(0, b)
                break





def checkClosed(n):
    
    global closedNodes
    
    for i in closedNodes:
        
        if n[3] == i[3]:
            return True
        
    return False

def checkOpen(n):
    
    global openNodes
    
    #if len(openNodes) > 10:
    
    #    for i in range(0, 10):
            
    #        if n[3] == openNodes[i][3]:
    #            return i
            
    #else:
        
    for i in range(0, len(openNodes)):
        
        if n[3] == openNodes[i][3]:
            return i
        
    return None
    
def checkObstacle(n):
    
    if canvas[n[3][1], n[3][0], 2] == 255:
        return True
        
    return False






def moveUp(n):
    
    global openNodes
    global ni
    
    ni += 1
    
    new = [n[0] + 1, ni, n[1], [n[3][0], n[3][1] - 1]]
    
    if (new[3][1] >= 0):
        
        if (checkObstacle(new) == False) and (checkClosed(new) == False):
        
            ex = checkOpen(new)
            
            if (ex == None):
                
                hq.heappush(openNodes, new)
                
            elif (new[0] < openNodes[ex][0]):
                
                openNodes.pop(ex)
                hq.heappush(openNodes, new)

def moveDown(n):
    
    global openNodes
    global ni
    
    ni += 1
    
    new = [n[0] + 1, ni, n[1], [n[3][0], n[3][1] + 1]]
    
    if (new[3][1] < 250):
        
        if (checkObstacle(new) == False) and (checkClosed(new) == False):
        
            ex = checkOpen(new)
            
            if (ex == None):
                
                hq.heappush(openNodes, new)
                
            elif (new[0] < openNodes[ex][0]):
                
                openNodes.pop(ex)
                hq.heappush(openNodes, new)

def moveLeft(n):
    
    global openNodes
    global ni
    
    ni += 1
    
    new = [n[0] + 1, ni, n[1], [n[3][0] - 1, n[3][1]]]
    
    if (new[3][0] >= 0):
        
        if (checkObstacle(new) == False) and (checkClosed(new) == False):
        
            ex = checkOpen(new)
            
            if (ex == None):
                
                hq.heappush(openNodes, new)
                
            elif (new[0] < openNodes[ex][0]):
                
                openNodes.pop(ex)
                hq.heappush(openNodes, new) 

def moveRight(n):
    
    global openNodes
    global ni
    
    ni += 1
    
    new = [n[0] + 1, ni, n[1], [n[3][0] + 1, n[3][1]]]
    
    if (new[3][0] < 250):
        
        if (checkObstacle(new) == False) and (checkClosed(new) == False):
        
            ex = checkOpen(new)
            
            if (ex == None):
                
                hq.heappush(openNodes, new)
                
            elif (new[0] < openNodes[ex][0]):
                
                openNodes.pop(ex)
                hq.heappush(openNodes, new)
                
def moveUpLeft(n):
    
    global openNodes
    global ni
    
    ni += 1
    
    new = [n[0] + 1.4, ni, n[1], [n[3][0] - 1, n[3][1] - 1]]
    
    if (new[3][1] >= 0) and (new[3][0] >= 0):
        
        if (checkObstacle(new) == False) and (checkClosed(new) == False):
        
            ex = checkOpen(new)
            
            if (ex == None):
                
                hq.heappush(openNodes, new)
                
            elif (new[0] < openNodes[ex][0]):
                
                openNodes.pop(ex)
                hq.heappush(openNodes, new)

def moveUpRight(n):
    
    global openNodes
    global ni
    
    ni += 1
    
    new = [n[0] + 1.4, ni, n[1], [n[3][0] + 1, n[3][1] - 1]]
    
    if (new[3][1] >= 0) and (new[3][0] < 250):
        
        if (checkObstacle(new) == False) and (checkClosed(new) == False):
        
            ex = checkOpen(new)
            
            if (ex == None):
                
                hq.heappush(openNodes, new)
                
            elif (new[0] < openNodes[ex][0]):
                
                openNodes.pop(ex)
                hq.heappush(openNodes, new)

def moveDownLeft(n):
    
    global openNodes
    global ni
    
    ni += 1
    
    new = [n[0] + 1.4, ni, n[1], [n[3][0] - 1, n[3][1] + 1]]
    
    if (new[3][1] < 250) and (new[3][0] >= 0):
        
        if (checkObstacle(new) == False) and (checkClosed(new) == False):
        
            ex = checkOpen(new)
            
            if (ex == None):
                
                hq.heappush(openNodes, new)
                
            elif (new[0] < openNodes[ex][0]):
                
                openNodes.pop(ex)
                hq.heappush(openNodes, new) 

def moveDownRight(n):
    
    global openNodes
    global ni
    
    ni += 1
    
    new = [n[0] + 1.4, ni, n[1], [n[3][0] + 1, n[3][1] + 1]]
    
    if (new[3][1] < 250) and (new[3][0] < 250):
        
        if (checkObstacle(new) == False) and (checkClosed(new) == False):
        
            ex = checkOpen(new)
            
            if (ex == None):
                
                hq.heappush(openNodes, new)
                
            elif (new[0] < openNodes[ex][0]):
                
                openNodes.pop(ex)
                hq.heappush(openNodes, new)
    
    


#Colors
white = (255, 255, 255)

#Polygon (pts)
hexagon = np.array([[300, 50], [365, 88], [365, 162], [300, 200], [235, 162], [235, 88]], np.int32)
hexagon = hexagon.reshape((-1, 1, 2))

triangle = np.array([[460, 25], [460, 225], [510, 125]], np.int32)
triangle = triangle.reshape((-1, 1, 2))

#New image (canvas)
canvas = np.zeros((250, 600, 3), dtype = "uint8")

#Obstacles (shapes)
cv.rectangle(canvas, (100, 0), (150, 100), white, -1)
cv.rectangle(canvas, (100, 150), (150, 250), white, -1)

cv.fillPoly(canvas, [hexagon], white)
cv.fillPoly(canvas, [triangle], white)

#cv.imshow("Canvas", canvas)
#cv.waitKey(0)


#closedList = {(cost, currentNode): parentNode}

#Start & end nodes
start = [80, 80]
end = [170, 80]
openNodes = []
hq.heapify(openNodes)
closedNodes = []
solution = []
ni = 0
l = 0

current = [0, 0, 0, [start[0], start[1]]]
closedNodes.append(current)

#hq.heappush(openNodes, current)

#print(openNodes)
#moveUp([2, 0, 0, [13, 12]])
#moveUp([1, 0, 0, [13, 17]])
#moveUp([5, 0, 0, [13, 18]])
#moveUp([7, 0, 0, [13, 19]])

#moveUp(current)
#moveDown(current)
#moveLeft(current)
#moveRight(current)

#print(openNodes)

#print(moveDown([0, 0, 0, [599, 248]]))
#print(moveLeft([0, 0, 0, [1, 0]]))
#print(moveRight([0, 0, 0, [598, 0]]))

#print(checkClosed([0, 0, 0, [5, 3]]))
#print(checkObstacle([0, 0, 0, [235, 89]]))

while (current[3] != end):
    
    l += 1
    #print(l)
    
    
    moveUp(current)
    moveDown(current)
    moveLeft(current)
    moveRight(current)
    moveUpLeft(current)
    moveUpRight(current)
    moveDownLeft(current)
    moveDownRight(current)
    
    current = hq.heappop(openNodes)
    closedNodes.append(current)
    
print("Found Solution!")
canvas[start[1], start[0]] = (3, 240, 252)
canvas[end[1], end[0]] = (0, 240, 10)
cv.imshow("Canvas", canvas)
cv.waitKey(5000)
    
for i in closedNodes:
    
    x = i[3][0]
    y = i[3][1]
    
    #print(i)
    #print(x, y)
    
    if i != closedNodes[0] and i != closedNodes[len(closedNodes) - 1]:
    
        canvas[y, x] = (0, 0, 200)
        cv.imshow("Canvas", canvas)
        cv.waitKey(1)
    
backtrack()

for i in solution:
    
    x = i[3][0]
    y = i[3][1]
    
    #print(i)
    #print(x, y)
    
    
    canvas[y, x] = (255, 2, 2)
    cv.imshow("Canvas", canvas)
    cv.waitKey(1)

print("DONE!")
    
cv.waitKey(10000)
cv.destroyAllWindows()
```
